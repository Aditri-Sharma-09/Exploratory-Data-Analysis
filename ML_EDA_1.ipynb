##Univariate Analysis

Bar plot for evaluating the count of wine with its quality rate

```python
quality_count = df['quality'].value_counts()
print(quality_count)
```

```python
plt.figure(figsize=(10,6))
plt.bar(quality_count.index, quality_count, color='pink')
plt.title("Count plot of quality")
plt.xlabel("Quality")
plt.ylabel("Count")
plt.show()
```

Swarm plot for showing outlier in data

```python
plt.figure(figsize=(10,6))
sns.swarmplot(x="quality", y="alcohol", data=df, palette='viridis')
plt.title('Swarm Plot for Quality and Alcohol')
plt.xlabel('Quality')
plt.ylabel('Alcohol')
plt.show()
```

##Bivariate Analysis

Pair plot for showing the distribution of individual variables

```python
sns.set_palette("Pastel1")
plt.figure(figsize=(10,6))
sns.pairplot(df)
plt.suptitle("Pair plot")
plt.show()
```

Violin plot for examining the relationship between alcohol and quality

```python
df['quality'] = df['quality'].astype(str)

plt.figure(figsize=(10,6))

sns.violinplot(x='quality', y='alcohol', data=df, palette={               '3': 'lightcoral', '4': 'lightblue', '5': 'lightgreen', '6': 'gold', '7': 'lightskyblue', '8': 'lightpink'}, alpha=0.7)

plt.title('Violin Plot for Quality and Alcohol')
plt.xlabel('Quality')
plt.ylabel('Alcohol')
plt.show()
```

Box Plot for examining the relationship between alcohol and Quality

```python
sns.boxplot(x='quality', y='alcohol', data=df)
```

##Multivariate Analysis

correlation matrix plot

```python
plt.figure(figsize=(10,6))
sns.heatmap(df.corr(), annot=True, cmap='coolwarm')

plt.title('Correlation Matrix')
plt.show()
```
